```python
# Import Libraries 

from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
#Connect to the Website

url = "https://www.amazon.in/gp/bestsellers/books"

headers = {'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/106.0.0.0 Safari/537.36'}
```

```python
#Pull in the Data

page = requests.get(url, headers = headers)

s1 = BeautifulSoup(page.content, 'html.parser')

books = s1.find_all(id="gridItemRoot")

print(books[0])
```

```python
# Retrieve the first book from the list of books.
book = books[0]
```

```python
#Retrieve the book's rank and additional information within a given book

rank = book.find('span', class_ = 'zg-bdg-text').text[1:]
print(rank)
```

```python
child = book.find('div', class_ = 'zg-grid-general-faceout').div
print(child)
```

```python
#Extract the title, author, and price information

title = child.contents[1].text
print(title)
```

```python
author = child.contents[2].text
print(author)
```

```python
price = child.contents[-1].text
print(price)
```

```python
#Dictionary initialization to store book's information
books_dict = {'rank':[], 'title':[], 'author':[], 'price':[]}

#Iterate through each book
for book in books:
    rank = book.find('span', class_ = 'zg-bdg-text').text[1:]
    books_dict['rank'].append(rank)
    
    child = book.find('div', class_ = 'zg-grid-general-faceout').div

    title = child.contents[1].text
    books_dict['title'].append(title)
    
    author = child.contents[2].text
    books_dict['author'].append(author)
    
    price = child.contents[-1].text
    books_dict['price'].append(price)
```

```python
#Dataframe for a structured and tabular representation of the book data
book_records = pd.DataFrame(books_dict)
```
